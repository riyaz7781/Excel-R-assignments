Importing the Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

Load the dataset

```python
data1=pd.read_csv("/content/ToyotaCorolla.csv",encoding="latin1")
data1.head()
```

Exploratory Data Analysis

```python
data1.info()
```

```python
data2=pd.concat([data1.iloc[:,2:4],data1.iloc[:,6:7],data1.iloc[:,8:9],data1.iloc[:,12:14],data1.iloc[:,15:18]],axis=1)
data2
```

```python
data2.isna().sum()
```

```python
data3=data2.rename({"Age_08_04":"Age","Quarterly_Tax":"Tax"},axis=1)
data3
```

```python
data3[data3.duplicated()]
```

```python
data4=data3.drop_duplicates().reset_index(drop=True)
data4
```

```python
data4.describe()
```

Correlation Analysis

```python
data4.corr()
```

Pair Plotting

```python
sns.set_style("darkgrid")
sns.pairplot(data4)
```

Model Building

```python
model=smf.ols("Price~Age+KM+HP+cc+Doors+Gears+Tax+Weight",data=data4).fit()
```

Model Testing

```python
model.params,model.summary()
```

```python
model.tvalues,model.pvalues
```

```python
model.rsquared,model.rsquared_adj
```

Building model for Insignficant values(CC and Doors)

```python
smf_cc=smf.ols("Price~cc",data=data4).fit()
smf_cc.tvalues,smf_cc.pvalues
smf_cc.summary()
```

```python
smf_doors=smf.ols("Price~Doors",data=data4).fit()
smf_doors.tvalues,smf_doors.pvalues
smf_doors.summary()
```

```python
smf_both=smf.ols("Price~cc+Doors",data=data4).fit()
smf_both.tvalues,smf_both.pvalues
smf_both.summary()
```

Residual Analysis

```python
sns.set_style("darkgrid")
sm.qqplot(model.resid,line="q")
plt.show()
```

```python
list(np.where(model.resid>6000))
```

```python
list(np.where(model.resid<-6000))
```

Homoscedacity Checking

```python
def plot_resid_values(val):
  return (val-val.mean())/val.std()
plt.scatter(plot_resid_values(model.fittedvalues),plot_resid_values(model.resid))
plt.title("Homoscedacity Check")
plt.show()
```

Model Deletion Diagnostics

1.Cooks Distance

```python
model_influence=model.get_influence()
(c,_)=model_influence.cooks_distance
fig=plt.figure(figsize=(8,8))
plt.stem(np.arange(len(data4)),np.round(c,2))
plt.show()
```

```python
np.argmax(c),np.max(c)
```

2.Influence plot

```python
fig,ax=plt.subplots(figsize=(10,6))
fig=influence_plot(model,ax=ax)
```

Leverage values

```python
k=data4.shape[1]
n=data4.shape[0]
leverage_off=(3*(k+1))/n
leverage_off
```

Improving the model
