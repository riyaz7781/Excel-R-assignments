Import the Necessary Libraries

```python
import pandas as pd
import numpy as np
import string
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
import warnings
warnings.filterwarnings("ignore")
```

Load the Dataset

```python
data1=pd.read_csv("/content/Elon_musk.csv",encoding="latin1",error_bad_lines = False,index_col=False)
data1
```

Dropping the Numerical Column

```python
data2=data1["Text"]
data2
```

Removing starting and last characters

```python
musk = [y.strip() for y in data2]
musk
```

Removes Empty String

```python
text=' '.join(musk)
text
```

```python
no_punc_text = text.translate(str.maketrans("\x92", "'", string.punctuation))
no_punc_text
```

Tokenisation

```python
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
text_tokens = word_tokenize(no_punc_text)
text_tokens
```

```python
text_tokens[0:20]
```

```python
len(text_tokens)
```

StopWords

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
my_stop_words = stopwords.words('english')
my_stop_words
```

```python
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:65])
```

```python
lower_words = [x.lower() for x in no_stop_tokens]
print(lower_words[0:45])
```

Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens[0:40])
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(' '.join(no_stop_tokens))
print(doc[0:40])
```

Lemmatisation

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas[0:22])
```

Feature Extraction

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(lemmas)
pd.DataFrame.from_records([vectorizer.vocabulary_]).T.sort_values(0,ascending=False).head(10)
```

```python
pd.DataFrame.from_records([vectorizer.vocabulary_]).T.sort_values(0,ascending=True).head(10)
```

```python
vectorizer.vocabulary_
```

```python
print(X.toarray()[50:100])
```

N-Grams (Bi-Grams,Tri-Grams)

```python
vectorizer_ngram_range = CountVectorizer(analyzer='word',
                                         ngram_range=(1,4),
                                         max_features = 100)

bow_matrix_ngram = vectorizer_ngram_range.fit_transform(musk)
bow_matrix_ngram
```

```python
vectorizer_ngram_range.get_feature_names_out()
```

TF-IDF Vectoriser

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer_n_gram_max_features = TfidfVectorizer(
                                                 analyzer='word',
                                                 ngram_range=(1,4),
                                                 max_features = 100)

tf_idf_matrix_n_gram_max_features = vectorizer_n_gram_max_features.fit_transform(musk)
```
