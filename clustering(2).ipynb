Importing the necessary libraries



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

Load the dataset

```python
data1=pd.read_csv("/content/crime_data.csv")
data1
```

Basic EDA

```python
data1.shape
```

```python
data1.isna().sum()
```

```python
data2=data1.rename({"Unnamed: 0":"States"},axis=1)
data2
```

Creating Function for Normalising the values

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return x
```

```python
data1_norm=norm_func(data1.iloc[:,1:])
data1_norm
```

HIERARCHICAL CLUSTERING

Using dendrogram,we can choose the no.of clusters

```python
dendrogram=sch.dendrogram(sch.linkage(data1_norm,method="average"))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="complete")
y_hc=hc.fit_predict(data1_norm)
y_hc
```

```python
data1["Clusters"]=hc.labels_
data1
```

```python
data1["Clusters"].value_counts()
```

K-MEANS

Here we can choose Elbow Method to choose K

```python
wcss=[]
for i in range(1,8):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(data1_norm)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,8),wcss)
plt.title("Elbow Curve")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(data1_norm)
model.labels_
```

```python
data1["Clusters"]=model.labels_
a=data1.drop(["Cluseters"],axis=1)
a
```

```python
a.groupby("Clusters").agg(["mean"]).reset_index()
```

DBSCAN

```python
data2=pd.read_csv("/content/crime_data.csv")
data3=data2.iloc[:,1:]
data3.values
```

```python
std=StandardScaler().fit(data3.values)
x=std.transform(data3.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
c=pd.DataFrame(dbscan.labels_,columns=["Clusters"])
c
```

```python
data4=pd.concat([data2,c],axis=1)
data4
```

```python
data4["Clusters"].value_counts()
```
